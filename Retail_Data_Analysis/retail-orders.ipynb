```python
#Treat "Not Available" and "unknown" as NULL 
df1=pd.read_csv(r"./orders.csv",na_values=['Not Available', 'unknown'])
df1['Ship Mode'].unique()
```

```python
#Rename columns, make them lower case and replace the space with underscore
#df1.rename(columns={'Order Id':'order_id'})
#1)
df1.columns=df.columns.str.lower()
df1.columns=df1.columns.str.replace(' ','_')
df1.columns
df1.head()
```

```python
#derive new columns :Discount, Sales price and profit
df1['discount']=df1['list_price']*df1['discount_percent']*.01
df1['sale_price']= df1['list_price']-df1['discount']
df1['profit']=df1['sale_price']-df1['cost_price']
df1
```

```python
#Convert "Order Date" from object datatype to Datetime
df1['order_date']=pd.to_datetime(df1['order_date'],format="%Y-%m-%d")
```

```python
#drop cost price list price and discount percent columns
df1.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
#load the data into sql server using replace option

import sqlalchemy as sal
engine = sal.create_engine('mssql://Satvik\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
#load the data into sql server using append option
df1.to_sql('df_orders', con=conn , index=False, if_exists = 'append')

#If used "append" ,Pandas will create a table with highest posssible datatype (varchar(max)...bigint..)
```
